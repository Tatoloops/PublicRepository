## Ejercicio 2: masa - resorte
Una masa de 4 kg se encuentra conectada a un resorte cuya constante de elasticidad es 1.6N/m. Además, el sistema se encuentra bajo la acción de una fuerza retardadora constante de amortiguamiento 10Ns/m. Además hay una fuerza externa f(t)= cos(6t) que actúa sobre este sistema. Si inicialmente el sistema se encuentra a 10 unidades a la izquierda de la posición de reposo, a una velocidad de 1m/s, encuentre la solución de EDO, y grafique el resultado.

Datos extraídos del enunciado:  
* 4 y'' + 10 y' + 1.6 y - cos(6t) = 0  
* y(0) = -10  
* y'(0) = 1  


```python
# Definir variables y función desconocida
t = sp.symbols('t')
y = sp.Function('y')

# Definir la ecuación diferencial
eq = 4*y(t).diff(t, t) + 10*y(t).diff(t) + 1.6*y(t) - sp.cos(6*t)

# Resolver la ecuación diferencial
sol = sp.dsolve(eq, y(t), ics={y(0): -10, y(t).diff(t).subs(t, 0): 1})

# Mostrar la solución
print("Solución de la EDO:")
sp.pprint(str(sol), use_unicode=True)

print("\nSolución de la EDO:")
sp.pprint(sol, use_unicode=True)

print("\nSolución de la EDO:")
print(sol)


# Graficar la solución
t_values = np.linspace(0, 50, 1000)
y_values = [sol.rhs.subs(t, val) for val in t_values]


plt.figure(figsize=(20, 10))
plt.plot(t_values, y_values)
plt.title('Solución de la EDO')
plt.xlabel('Tiempo \'t\' (segundos)')
plt.ylabel('Posición resorte \'y\' (metros)')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 10))
plt.plot(t_values, y_values)
plt.title('Solución de la EDO')
plt.xlabel('Tiempo \'t\' (segundos)')
plt.ylabel('Posición resorte \'y\' (metros)')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 10))
plt.plot(t_values, y_values)
plt.xlim(0, 2)
plt.ylim(-10, -8)
plt.title('Solución de la EDO')
plt.xlabel('Tiempo \'t\' (segundos)')
plt.ylabel('Posición resorte \'y\' (metros)')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 10))
plt.plot(t_values, y_values)
plt.xlim(28, 30)
plt.ylim(-1.5, 0.5)
plt.title('Solución de la EDO')
plt.xlabel('Tiempo \'t\' (segundos)')
plt.ylabel('Posición resorte \'y\' (metros)')
plt.grid(True)
plt.show()
```
