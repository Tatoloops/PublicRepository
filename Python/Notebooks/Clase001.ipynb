### Joins

Podemos hacer _joins_ sobre los Data Frames. Partamos con un ejemplo sencillo de dos objetos de tipo `DataFrame` que comparten el nombre de un atributo en el que se desea hacer _join_:

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': np.arange(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd', 'a'],
                    'data2': np.arange(8, 12)})

pd.merge(df1, df2)
```

```python
df1
```

En la operación anterior, omitimos indicar explícitamente el atributo sobre el que estamos haciendo join. Para indicarlo hacemos lo siguiente:

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': np.arange(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd', 'a'],
                    'data2': np.arange(8, 12)})

pd.merge(df1, df2, on='key')
```

En el caso de que los atributos se llamen de distinta forma:

```python
df1 = pd.DataFrame({'key1': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': np.arange(7)})
df2 = pd.DataFrame({'key2': ['a', 'b', 'd', 'a'],
                    'data2': np.arange(8, 12)})

pd.merge(df1, df2, left_on='key1', right_on='key2')
```

En el caso de necesitar un _Outer Join_ (esto es, seguir incluyendo valores que no hacen _match_ con alguna fila en la otra tabla), podemos indicarlo con el argumento `how`. Puede ser `'left'`, `'right'` o `'outer'`. Veamos un ejemplo de _Left Outer Join_:

```python
df1 = pd.DataFrame({'key1': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': np.arange(7)})
df2 = pd.DataFrame({'key2': ['a', 'b', 'd', 'a'],
                    'data2': np.arange(8, 12)})

pd.merge(df1, df2, left_on='key1', right_on='key2', how='left')
```

En el caso de querer un _join_ por más de un argumento, puedo indicar una lista de atributos. También puedes renombrar atributos en el caso de que su nombre sea igual en ambos Data Frame y no quieras generar conflictos. Esto lo puedes hacer mediante el argumento `suffixes`. Para ver más puedes consultar la documentación.

Lo último que veremos es cómo hacer un _join_ utilizando una de los índices.

## **1.6 Ejemplos prácticos de algoritmos recursivos**

**Factorial:**
El factorial de un número n (denotado como n!) es el producto de todos los números enteros positivos desde 1 hasta n. Se puede calcular de manera recursiva de la siguiente manera:

```python
n=int(input("Ingrese el valor de n: "))

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

resultado=factorial(n)
print(f"El factorial de {n} es {resultado}")
```

**Fibonacci:**
La secuencia de Fibonacci es una secuencia de números en la que cada número es la suma de los dos anteriores. Se puede calcular de manera recursiva de la siguiente manera:

```python
n=int(input("Ingrese el valor de n: "))

def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

resultado = fibonacci(n)
print(f"El término {n}-ésimo de la secuencia de Fibonacci es {resultado}")
```

**Búsqueda binaria recursiva:**
La búsqueda binaria es un algoritmo que se utiliza para encontrar la posición de un elemento en un arreglo ordenado. Puede implementarse de manera recursiva de la siguiente manera:

```python
mi_arreglo = [1, 3, 5, 7, 9, 11, 13, 15]
```

```python
elemento_buscar=int(input("Ingrese el elemento a buscar: "))

def busqueda_binaria(arr,elemento, inicio=0, fin=None):
    if fin is None:
        fin = len(arr) - 1
    if inicio > fin:
        return - 1
    medio = (inicio + fin)// 2

    if arr[medio] == elemento:
        return medio
    elif arr[medio] < elemento:
        return busqueda_binaria(arr,elemento, medio + 1, fin)
    else:
        return busqueda_binaria(arr,elemento, inicio, medio - 1)

resultado = busqueda_binaria(mi_arreglo, elemento_buscar)
if resultado != -1:
    print(f"El elemento {elemento_buscar} se encuentra en la posición {resultado+1}")
else:
    print(f"El elemento {elemento_buscar} no se encuentra en el arreglo")
```

Implementa una función recursiva para calcular la suma de los primeros n números naturales:

```python
n=int(input("Ingrese el valor de n: "))

def suma_naturales(n):
    if n <= 0:
        return 0
    else:
        return n + suma_naturales(n-1)

resultado = suma_naturales(n)
print(f"La suma de los primeros {n} números naturales es {resultado}")
```

Implementa una función recursiva para calcular la potencia de un número:

```python
base=int(input("Ingrese la base: "))
exponente=int(input("Ingrese el exponente: "))

def potencia(base, exponente):
    if exponente == 0:
        return 1
    else:
        return base * potencia(base, exponente  - 1)
resultado = potencia(base, exponente)
print(f"{base}^{exponente} = {resultado}")
```

## **1.7 Parámetros y Argumentos en Python**

**1. Uso de parámetros posicionales y con nombre:**
En Python, las funciones pueden recibir argumentos de dos formas: mediante parámetros posicionales y con nombre. Los parámetros posicionales son aquellos que se pasan en el orden en que fueron definidos en la función. Los parámetros con nombre, también conocidos como argumentos con nombre o keywords, se identifican mediante el nombre del parámetro seguido de un signo igual (=) y el valor correspondiente.

```python
def saludar(nombre, edad):
    print(f"Hola, me llamo {nombre} y tengo {edad} años")

saludar("Juan", 30)
```

```python
def saludar(nombre, edad):
    print(f"Hola, me llamo {nombre} y tengo {edad} años")

saludar(edad=30, nombre="Juan")
```

```python
def saludar(nombre, edad=25):
    print(f"Hola, me llamo {nombre} y tengo {edad} años")

saludar("Juan")
```

En este caso, si no se proporciona un valor para edad, se utilizará el valor predeterminado de 25. Sin embargo, aún puede especificar un valor diferente si es necesario utilizando argumentos con nombre. Por ejemplo: saludar("Juan", edad=30).

**2. Funciones con un número variable de argumentos:**
En Python, es posible definir funciones que acepten un número variable de argumentos. Para hacerlo, se utilizan los operadores `*args` y `kwargs`. El operador `*args` permite pasar una cantidad arbitraria de argumentos posicionales como una tupla, y el operador `kwargs` permite pasar una cantidad arbitraria de argumentos con nombre como un diccionario.

Operador *args (argumentos posicionales):
El operador *args permite que una función acepte una cantidad arbitraria de argumentos posicionales y los agrupa en una tupla. Esto significa que puedes pasar cualquier número de argumentos posicionales cuando llamas a la función, y se recogerán en forma de tupla dentro de la función. Aquí tienes un ejemplo:

```python
def suma(*args):
    resultado = 0
    for numero in args:
        resultado += numero
    return resultado

# Llamamos a la función con diferentes cantidades de argumentos
print(suma(1, 2))             # Imprime 3
print(suma(1, 2, 3, 4, 5))    # Imprime 15
```

Operador **kwargs (argumentos con nombre):
El operador **kwargs permite que una función acepte una cantidad arbitraria de argumentos con nombre y los agrupa en un diccionario. Esto significa que puedes pasar una serie de argumentos con nombre cuando llamas a la función, y se recogerán en forma de diccionario dentro de la función. Aquí tienes un ejemplo:

```python
def imprimir_datos(**kwargs):
    for clave, valor in kwargs.items():
        print(f"{clave}: {valor}")

# Llamamos a la función con diferentes argumentos con nombre
imprimir_datos(nombre="Juan", edad=30, ciudad="Nueva York")
```

En este caso, la función imprimir_datos acepta una serie de argumentos con nombre y los imprime en forma de pares clave-valor.

En resumen, los operadores *args y **kwargs son herramientas poderosas para crear funciones flexibles que pueden manejar una cantidad variable de argumentos posicionales y con nombre. Esto es especialmente útil cuando trabajas con funciones genéricas o cuando deseas proporcionar opciones adicionales sin tener que modificar constantemente la firma de la función.

**3. Argumentos predeterminados:**
En Python, es posible asignar valores predeterminados a los parámetros de una función. Si se llama a la función sin proporcionar un valor para un parámetro con valor predeterminado, este tomará el valor asignado en la definición de la función.

```python
def saludar(nombre, mensaje="Hola"):
    print(f"{mensaje}, {nombre}!")

# Llamamos a la función sin proporcionar un valor para 'mensaje'
saludar("Juan")  # Imprime "Hola, Juan!"

# Llamamos a la función proporcionando un valor para 'mensaje'
saludar("Maria", "Buenos días")  # Imprime "Buenos días, Maria!"
```

En el primer llamado a saludar, no proporcionamos un valor para mensaje, por lo que la función utiliza el valor predeterminado "Hola". En el segundo llamado, proporcionamos un valor diferente para mensaje, y la función utiliza ese valor en lugar del predeterminado.

Los parámetros con valores predeterminados son útiles cuando deseas que una función tenga un comportamiento estándar, pero también quieres dar la opción a los usuarios de la función de personalizar ese comportamiento si lo desean. Esto hace que tus funciones sean más flexibles y fáciles de usar en diferentes situaciones sin necesidad de crear múltiplas versiones de la misma función.

Lo primero que quisieramos configurar es el rango de los ejes. Podemos hacer esto usando  `ylim` y `xlim`, o `axis('tight')` para obtener automáticamente rangos de ejes "apretados".

### La API orientada al objeto de Matplotlib

La idea principal de la programación orientada al objeto es que se tienen objetos sobre los cuales se pueden aplicar funciones y acciones, y que ningún estado de un objeto o programa debe ser global (como en pyplot). La ventaja de esta forma de trabajo se manifiesta cuando se requiere crear más de una figura, o cuando una figura contiene sub-figuras.

Para usar la API orientada al objeto comenzamos similarmente al ejemplo anterior, pero en lugar de crear una instancia de figura global almacenamos una referencia a la figura recién creada en la variable `fig`, y a partir de ella creamos nuevos ejes `axes` usando el método `add_axes` en la instancia `fig` de la clase `Figure`.

```python
fig = plt.figure()

ejes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # izquierda, abajo, ancho, altura (rango 0 a 1)

ejes.plot(x,y, color = 'r')

ejes.set_xlabel('x')
ejes.set_ylabel('y')
ejes.set_title(u'Título');
```

Aunque se requiere algo más de código, la ventaja es que ahora tenemos control completo sobre dónde se ubican los ejes, y además podemos agregar fácilmente más de un eje a la figura.

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # ejes principales
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # ejes del gráfico insertado

# figura principal
axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title(u'Título principal')

# figura insertada
axes2.plot(y,x,'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title(u'Título secundario');
```

Si no nos importa ser explícitos sobre dónde estarán ubicados los ejes en el marco de nuestra figura, podemos usar uno de los muchos administradores de la distribución de los ejes que tiene Matplotlib. Un favorito es `subplots`, que puede ser usado de la forma siguiente:

```python
fig, axes = plt.subplots() # crea una nueva figura y un set de ejes

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title(u'Título');
```

```python
 # crea una figura, con dos set de ejes, almacenados en el array axes
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(u'Título');
```

Fácil, pero no tan bonito debido a la superposición de los ejes y etiquetas en la parte central, cierto?

Podemos mejorar esto usando el método `fig.tight_layout`, que ajusta automáticamente la posición de los ejes en el marco de la figura de modo que no exista contenido que se superponga:

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(u'Título')

fig.tight_layout()
```
